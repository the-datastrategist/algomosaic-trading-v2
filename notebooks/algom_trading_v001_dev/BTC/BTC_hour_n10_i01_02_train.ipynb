## BTCUSD | ROR_n10


```python
# Load model data
from data import dataBlob
data = dataBlob("""
select * 
from `algomosaic-nyc.algom_dev_features.features_BTC_USD_hour_i01_201912`
""")
data.df = data.df.dropna()
data.df.head()
```

```python
# Initialize model
from model_regression import modelRegression
model = modelRegression(
    data, 
    outcome='ROR_n10', 
    index_features=configs.index_features, 
    omit_features=configs.omit_features
)

# Specify model parameters
from sklearn import ensemble
reg = ensemble.GradientBoostingRegressor(
    loss='ls', 
    learning_rate=0.1,
    n_estimators=800,
    subsample=.9,
    criterion='friedman_mse', 
    min_samples_split=2, 
    min_samples_leaf=1, 
    min_weight_fraction_leaf=0.0, 
    max_depth=3,
    min_impurity_decrease=0.0, 
    min_impurity_split=None, 
    init=None,
    random_state=None, 
    max_features=None, 
    alpha=0.9, 
    verbose=0,
    max_leaf_nodes=None, 
    warm_start=False, 
    validation_fraction=0.1, 
    n_iter_no_change=None, 
    tol=0.0001
)

# Train model
model.train(reg)
```

```python
# Model IDs
print('model_id: ' + model.model_id)
print('model_execution_id: ' + model.model_execution_id)
```

```python
# Save model (optional)
model.save()
```
