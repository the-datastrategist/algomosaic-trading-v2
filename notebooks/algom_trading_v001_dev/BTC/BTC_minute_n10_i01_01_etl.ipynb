### BTC-USD -- minute -- i01 -- YYYYMM

```python
model = ticker_etl.run_etl_process(
    ticker='BTC-USD',
    start_date='2020-01-01',
    end_date='2020-02-01',
    project='algomosaic-nyc',
    destination_table='algom_dev_features.features_{ticker}_{interval}_{iteration}_20200101',
    table_params={
        'ticker': 'BTC-USD',
        'interval': 'minute',
        'iteration': 'i01'
    },
    interval='minute',
    exchange='CCCAGG',
    data_library='etl.cryptocompare_ticker_data',
    features_library='etl.algom_trading_v001.get_features_minute_i01',
    to_bq=True
)
model.data.df.tail()
```
